```python
from PIL import Image
```

```python
image = Image.open(r'/home/riya-gupta/Documents/personal docs/Riya-passport.jpg')
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(image)
```

```python
image.format
```

```python
image.size      # (x,y) = width * height
```

```python
image.mode
```

```python
# CONVERT RGB TO GRAY IMAGE
from PIL import ImageOps
```

```python
image_gray = ImageOps.grayscale(image)
```

```python
plt.imshow(image_gray)
```

```python
image_gray.mode            #GRAY SCALE
```

```python
image_gray.size               #Definitely SIZE remains same after conversion
```

```python
image_gray.format           #Not saved in any format
```

```python
#Save image to .jpg format
image_gray.save("riya_image_gray.jpg")
```

```python
#Change the intensity of the image 
image_gray.quantize(10)
```

```python
image.quantize(200)
```
